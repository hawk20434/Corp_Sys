Задание 1.1.3. Проверить качество данных (например, отсутствующие значения и выбросы).

```python
# посмотрим пропущенные значения
missing_values = df.isnull().sum()
missing_values
```

```python
# процент пропущенных значений
mysize = df.index.size
missing_count = ((missing_values / mysize) * 100)
missing_count
```

```python
# запуск вычисления, используя метод подсчета
with ProgressBar():
  missing_count_percent = missing_count.compute()
missing_count_percent
```

Задание 1.2.1. Визуализировать DAG с одним узлом и зависимостями.

```python
import dask.delayed as delayed
```

```python
def increment(i):
    return i + 1

def add(x, y):
    return x + y

# функции переноса внутри отложенного объекта и цепочки
x = delayed(increment)(1)
y = delayed(increment)(2)
z = delayed(add)(x, y)

# visualize the DAG
z.visualize()
```

```python
# show the result
z.compute()
```

Задание 1.2.1. Визуализировать DAG с более чем одним узлом и зависимостями.

```python
data = [1, 2, 3, 4, 5]

# compile first layer and visualize
layer1 = [delayed(increment)(i) for i in data]
total1 = delayed(sum)(layer1)
total1.visualize()
```

```python
def double(x):
    return x * 2

# compile second layer and visualize
layer2 = [delayed(double)(j) for j in layer1]
total2 = delayed(sum)(layer2)#.persist()
total2.visualize()
```

```python
z = total2.compute()
z
```
