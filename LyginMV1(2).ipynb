```python
df.drop(['EmployeeCount', 'EmployeeNumber', 'Over18', 'StandardHours'], axis="columns", inplace=True)
df.head()
```

–î–ª—è –Ω–∞—á–∞–ª–∞ –Ω–µ–æ–±—Ö–æ–¥–∏–º–æ –æ–∑–Ω–∞–∫–æ–º–∏—Ç—å—Å—è –ø–æ–¥—Ä–æ–±–Ω–µ–µ —Å –¥–∞–Ω–Ω—ã–º–∏, –ø–æ–Ω—è—Ç—å —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –∏ –≤—ã–¥–≤–∏–Ω—É—Ç—å –≥–∏–ø–æ—Ç–µ–∑—ã. –í–æ–∑—å–º—É —Å–∞–º—ã–µ –ø—Ä–æ—Å—Ç—ã–µ –¥–ª—è –ø–µ—Ä–≤–∏—á–Ω–æ–π –∏–Ω—Ç–µ—Ä–ø—Ä–∏—Ç–∞—Ü–∏–∏ –¥–∞–Ω–Ω—ã–µ. –¢–∞–∫ –∫–∞–∫ –Ω–∞ –¥–∞–Ω–Ω–æ–º —ç—Ç–∞–ø–µ —Ç—Ä–µ–±—É–µ—Ç—Å—è —É—Å—Ç–∞–Ω–æ–≤–∏—Ç—å, —á—Ç–æ –≤–ª–∏—è–µ—Ç –Ω–∞ –ø–æ–∫–∞–∑–∞—Ç–µ–ª—å  - Attrition.

```python
# –ò–∑–º–µ–Ω–∏–º –¥–ª—è —É–¥–æ–±—Å—Ç–≤–∞ –∏ —Ä–∞—Å—á–µ—Ç–∞ –≤ –± –±—É–¥—É—â–µ–º
label = LabelEncoder()
df["Attrition"] = label.fit_transform(df.Attrition)
# –†–∞–∑–¥–µ–ª–∏–º –¥–∞–Ω–Ω—ã–µ.
attrition_yes = df[df['Attrition'] == 1]
attrition_no = df[df['Attrition'] == 0]
```

```python
plt.figure(figsize=(15, 6))
plt.hist(attrition_yes['Age'], alpha=0.5, label='Attrition: Yes', color='r')
plt.hist(attrition_no['Age'], alpha=0.5, label='Attrition: No', color='g')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Attrition - Age')
plt.legend()
plt.show()
```

–í–æ–∑—Ä–∞—Å—Ç –Ω–µ –ø–æ–º–æ–≥ —Å—Ä–∞–∑—É –≤—ã—è–≤–∏—Ç—å –∫–∞–∫—É—é —Ç–æ –≤–∑–∞–∏–º–æ—Å–≤—è–∑—å. –í–µ—Ä–æ—è—Ç–Ω–æ –µ—Å–ª–∏ –±—ã —Å—Ä–µ–¥–Ω–∏–π –≤–æ–∑—Ä–∞—Å—Ç –±—ã–ª –∑–Ω–∞—á–∏—Ç–µ–ª—å–Ω–æ –≤—ã—à–µ - —Ç–æ k—é–¥–∏ –±–æ–ª—å—à–µ –±—ã–ª–∏ –Ω–µ–¥–æ–≤–æ–ª—å–Ω—ã –∏ –º–µ–Ω—è–ª–∏ —Ä–∞–±–æ—Ç—É. –î–≤–∞ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –≤–∏–¥–Ω—ã, —á—Ç–æ —Å–µ–º–µ—Ç—Ä–∏—á–Ω—ã –∑–∞ –∏—Å–∫–ª—é—á–µ–Ω–∏–µ–º –æ–±—å–µ–º–∞ –≤—ã–±–æ—Ä–∫–∏. \\
–ü–æ—Å–º–æ—Ç—Ä—é —Ç–µ–ø–µ—Ä—å —á–∞—Å—Ç–æ—Ç—É –ø–æ–µ–∑–¥–æ–∫, –æ–Ω–∞ –≤–ø–æ–ª–Ω–µ –º–æ–∂–µ—Ç –≤–ª–∏—è—Ç—å –Ω–∞ —ç—Ç–æ.

```python
plt.figure(figsize=(10, 4))
plt.hist(attrition_yes['BusinessTravel'], bins=5, alpha=0.5, label='Attrition: Yes', color='r')
plt.hist(attrition_no['BusinessTravel'], bins=5, alpha=0.5, label='Attrition: No', color='g')
plt.xlabel('BusinessTravel')
plt.ylabel('Frequency')
plt.title('Attrition - BusinessTravel')
plt.legend()
plt.show()
```

–ß–∞—Å—Ç–æ—Ç–∞ –ø–æ–µ–∑–¥–æ–∫, —Ç–æ–∂–µ –Ω–µ –ø–æ–∫–∞–∑–∞–ª–∞ –∫–∞–∫–∏—Ö —Ç–æ —è–≤–Ω—ã—Ö –≤–ª–∏—è–Ω–∏–µ —ç—Ç–æ–≥–æ —Å—Ç–æ–ª–±—Ü–∞.

–ü–æ—Å–ª–µ–¥–Ω–µ–µ –ø—Ä–µ–¥–ø–æ–ª–æ–∂–µ–Ω–∏–µ, –∫–æ—Ç–æ—Ä–æ–µ —Å—Ä–∞–∑—É –ø—Ä–∏—Ö–æ–¥–∏—Ç –Ω–∞ —É–º -  –¥–∏—Å—Ç–∞–Ω—Ü–∏—è –¥–æ –¥–æ–º–∞ –º–æ–∂–µ—Ç –≤–ª–∏—è—Ç—å. –ü–æ—Å–º–æ—Ç—Ä–∏–º:

```python
plt.figure(figsize=(10, 4))
plt.hist(attrition_yes['DistanceFromHome'], alpha=0.5, label='Attrition: Yes', color='r')
plt.hist(attrition_no['DistanceFromHome'], alpha=0.5, label='Attrition: No', color='g')
plt.xlabel('Distance from Home')
plt.ylabel('Frequency')
plt.title('Attrition - Distance from Home')
plt.legend()
plt.show()
```

–¢–∞–∫–∂–µ –Ω–µ –±—ã–ª–æ –≤—ã—è–≤–ª–µ–Ω–æ –∑–Ω–∞—á–∏–º—ã—Ö –∑–∞–∫–æ–Ω–æ–º–µ—Ä–Ω–æ—Å—Ç–µ–π, –ø–µ—Ä–µ–±–∏—Ä–∞—Ç—å –≤—Å–µ –ø–∞—Ä–∞–º–µ—Ç—Ä—ã –±—É–¥–µ—Ç –Ω–µ—ç—Ñ—Ñ–µ–∫—Ç–∏–≤–Ω–æ, –ø–æ—ç—Ç–æ–º—É –≤–æ—Å–ø–æ–ª—å–∑—É–µ–º—Å—è –¥—Ä—É–≥–∏–º –º–µ—Ç–æ—Ä–¥–æ–º –¥–ª—è –≤—ã—è–≤–ª–µ–Ω–∏—è —É—Ö–æ–¥–∞-–æ—Ç—Ç–æ–∫–∞ –∏–∑ –∫–æ–º–ø–∞–Ω–∏–∏. –í–æ—Å–ø–æ–ª—å–∑—É–µ–º—Å—è –∫–æ–æ—Ä—Ä–µ–ª—è—Ü–∏–æ–Ω–Ω–æ–π –º–∞—Ç—Ä–∏—Ü–µ–π.

```python
col = df.corr().nlargest(20, "Attrition").Attrition.index
plt.figure(figsize=(10, 10))
sns.heatmap(df[col].corr(), annot=True, cmap="RdYlGn", annot_kws={"size":7}, fmt=".2f")
```
