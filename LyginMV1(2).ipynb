'EmployeeCount', 'Over18', 'StandardHours' имеют только одно уникальное значение - поэтому для анализа его лучше исключить (количество колонок и так довольно большое, что усложняет анализ). Эти колонки не содержат полезной информации
'EmployeeNumber' имеет 1470 уникальных значений - показывает просто уникальный номер сотрудника, что логично, но данный столбец, также мало что может сказать о данных кроме их количества.

```python
df.drop(['EmployeeCount', 'EmployeeNumber', 'Over18', 'StandardHours'], axis="columns", inplace=True)
df.head()
```

Для начала необходимо ознакомиться подробнее с данными, понять распределения и выдвинуть гипотезы. Возьму самые простые для первичной интерпритации данные. Так как на данном этапе требуется установить, что влияет на показатель  - Attrition.

```python
# Изменим для удобства и расчета в б будущем
label = LabelEncoder()
df["Attrition"] = label.fit_transform(df.Attrition)
# Разделим данные.
attrition_yes = df[df['Attrition'] == 1]
attrition_no = df[df['Attrition'] == 0]
```

```python
plt.figure(figsize=(15, 6))
plt.hist(attrition_yes['Age'], alpha=0.5, label='Attrition: Yes', color='r')
plt.hist(attrition_no['Age'], alpha=0.5, label='Attrition: No', color='g')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Attrition - Age')
plt.legend()
plt.show()
```

Возраст не помог сразу выявить какую то взаимосвязь. Вероятно если бы средний возраст был значительно выше - то kюди больше были недовольны и меняли работу. Два распределения видны, что семетричны за исключением обьема выборки. \\
Посмотрю теперь частоту поездок, она вполне может влиять на это.

```python
plt.figure(figsize=(10, 4))
plt.hist(attrition_yes['BusinessTravel'], bins=5, alpha=0.5, label='Attrition: Yes', color='r')
plt.hist(attrition_no['BusinessTravel'], bins=5, alpha=0.5, label='Attrition: No', color='g')
plt.xlabel('BusinessTravel')
plt.ylabel('Frequency')
plt.title('Attrition - BusinessTravel')
plt.legend()
plt.show()
```

Частота поездок, тоже не показала каких то явных влияние этого столбца.

Последнее предположение, которое сразу приходит на ум -  дистанция до дома может влиять. Посмотрим:

```python
plt.figure(figsize=(10, 4))
plt.hist(attrition_yes['DistanceFromHome'], alpha=0.5, label='Attrition: Yes', color='r')
plt.hist(attrition_no['DistanceFromHome'], alpha=0.5, label='Attrition: No', color='g')
plt.xlabel('Distance from Home')
plt.ylabel('Frequency')
plt.title('Attrition - Distance from Home')
plt.legend()
plt.show()
```

Также не было выявлено значимых закономерностей, перебирать все параметры будет неэффективно, поэтому воспользуемся другим метордом для выявления ухода-оттока из компании. Воспользуемся коорреляционной матрицей.

```python
col = df.corr().nlargest(20, "Attrition").Attrition.index
plt.figure(figsize=(10, 10))
sns.heatmap(df[col].corr(), annot=True, cmap="RdYlGn", annot_kws={"size":7}, fmt=".2f")
```

**Можно сделать следующие выводы по данным:**

1.   Пропущенные значения не были выявлены.
2.   Были обнаружены ряд колонок, которые не являются ценными для анализа и их стоит удалить из выборки.

**Можно сделать следующие выводы из визуального и матриченого анализа:**

2.   Месячный доход высоко коррелирует с общим количеством рабочих часов.
3.   Возраст также положительно коррелирует с общим количеством рабочих часов.
